ExplicaciÃ³n de las columnas y valores Ãºnicos del dataset TelecomX

---

*Customer*:

1. **gender**: GÃ©nero del cliente.  
   - `'Female'`: Mujer  
   - `'Male'`: Hombre

2. **SeniorCitizen**: Indica si el cliente es adulto mayor.  
   - `0`: No es adulto mayor (< 65 aÃ±os)  
   - `1`: Es adulto mayor (â‰¥ 65 aÃ±os)

3. **Partner**: Indica si el cliente tiene pareja.  
   - `'Yes'`: Tiene pareja  
   - `'No'`: No tiene pareja

4. **Dependents**: Indica si el cliente tiene personas a su cargo (hijos, familiares).  
   - `'Yes'`: Tiene dependientes  
   - `'No'`: No tiene dependientes

5. **tenure**: Tiempo (en meses) que el cliente lleva con la compaÃ±Ã­a.  
   - Valores enteros desde 0 hasta mÃ¡s de 70 meses, indicando antigÃ¼edad.

---

*Phone*:

1. **PhoneService**: Indica si el cliente tiene contratado servicio telefÃ³nico fijo con la compaÃ±Ã­a.  
   - `'Yes'`: Tiene servicio telefÃ³nico  
   - `'No'`: No tiene servicio telefÃ³nico

2. **MultipleLines**: Indica si el cliente tiene mÃºltiples lÃ­neas telefÃ³nicas bajo el mismo contrato (aplica solo si PhoneService es `'Yes'`).  
   - `'No'`: Solo una lÃ­nea telefÃ³nica  
   - `'Yes'`: MÃ¡s de una lÃ­nea telefÃ³nica  
   - `'No phone service'`: No tiene servicio telefÃ³nico (equivale a PhoneService = 'No')

---

*Internet*:

1. **InternetService**: Tipo de servicio de internet contratado.  
   - `'DSL'`: Servicio por lÃ­nea DSL  
   - `'Fiber optic'`: Servicio por fibra Ã³ptica  
   - `'No'`: No tiene servicio de internet

2. **OnlineSecurity**: Servicio adicional de seguridad en lÃ­nea.  
   - `'Yes'`: Tiene contratado  
   - `'No'`: No tiene contratado  
   - `'No internet service'`: No tiene internet (por tanto, no aplica)

3. **OnlineBackup**: Servicio adicional de respaldo en lÃ­nea (backup).  
   - `'Yes'`: Tiene contratado  
   - `'No'`: No tiene contratado  
   - `'No internet service'`: No tiene internet

4. **DeviceProtection**: Servicio adicional de protecciÃ³n para dispositivos (antivirus, seguro, etc).  
   - `'Yes'`: Tiene contratado  
   - `'No'`: No tiene contratado  
   - `'No internet service'`: No tiene internet

5. **TechSupport**: Servicio de soporte tÃ©cnico premium.  
   - `'Yes'`: Tiene contratado  
   - `'No'`: No tiene contratado  
   - `'No internet service'`: No tiene internet

6. **StreamingTV**: Servicio de televisiÃ³n por streaming.  
   - `'Yes'`: Tiene contratado  
   - `'No'`: No tiene contratado  
   - `'No internet service'`: No tiene internet

7. **StreamingMovies**: Servicio de pelÃ­culas por streaming.  
   - `'Yes'`: Tiene contratado  
   - `'No'`: No tiene contratado  
   - `'No internet service'`: No tiene internet

---

*Account*:

1. **Contract**: Tipo de contrato del cliente.  
   - `'One year'`: Contrato de un aÃ±o  
   - `'Month-to-month'`: Contrato mes a mes (sin compromiso a largo plazo)  
   - `'Two year'`: Contrato de dos aÃ±os

2. **PaperlessBilling**: Indica si el cliente recibe facturas electrÃ³nicas (sin papel).  
   - `'Yes'`: Factura electrÃ³nica  
   - `'No'`: Factura en papel

3. **PaymentMethod**: MÃ©todo que usa el cliente para pagar sus facturas.  
   - `'Mailed check'`: Pago mediante cheque enviado por correo  
   - `'Electronic check'`: Pago electrÃ³nico mediante cheque  
   - `'Credit card (automatic)'`: Pago automÃ¡tico con tarjeta de crÃ©dito  
   - `'Bank transfer (automatic)'`: Pago automÃ¡tico mediante transferencia bancaria

4. **Charges.Monthly**: Monto mensual que paga el cliente (en moneda local o USD).

5. **Charges.Total**: Monto total acumulado que ha pagado el cliente desde el inicio del servicio (puede estar en formato texto).

Se puede ver que *tenure* es 0 en todos los espacios vacÃ­os, por lo que en vez de NaN lo reemplazamos por 0, ya que no tiene antigÃ¼edad.

```python
# Reemplazar cadenas vacÃ­as o con solo espacios por NaN
df_final['account.Charges.Total'] = df_final['account.Charges.Total'].astype(str).str.strip().replace('', pd.NA)

# Convertir la columna a numÃ©rica (si no lo es ya), forzando errores a NaN
df_final['account.Charges.Total'] = pd.to_numeric(df_final['account.Charges.Total'], errors='coerce')

# Reemplazar NaN por 0
df_final['account.Charges.Total'] = df_final['account.Charges.Total'].fillna(0)

df_final['account.Charges.Total'].isna().sum()  # DeberÃ­a dar 0
```

Se realizÃ³ un ETL sobre los datos de clientes, servicios e informaciÃ³n de contrato, aplicando limpieza de datos, normalizaciÃ³n de columnas binarias, correcciÃ³n de valores faltantes y cÃ¡lculo de mÃ©tricas como cargos diarios. Durante la limpieza se detectaron inconsistencias: 224 filas de Churn con strings vacÃ­os fueron eliminadas (3,08% del total), y 11 filas con cargos totales NaN fueron corregidas, ya que la informaciÃ³n podÃ­a inferirse.

Del anÃ¡lisis de los datos limpios, respaldado por los grÃ¡ficos generados (histograma de churn por tenure, lÃ­nea de tasa de abandono por antigÃ¼edad, boxplots y scatter de cargos mensuales, y grÃ¡ficos categÃ³ricos por gÃ©nero, mÃ©todo de pago, streaming, servicio de internet y tipo de contrato), se concluye que el abandono total de clientes alcanza el 26,5%, concentrÃ¡ndose principalmente en contratos mes a mes, clientes con fibra Ã³ptica y pagos electrÃ³nicos mediante cheques. La mayor tasa de abandono ocurre durante los primeros 6 meses de antigÃ¼edad, disminuyendo significativamente a medida que los clientes permanecen mÃ¡s tiempo. Los clientes que abandonaron presentaron una mediana de cargos mensuales de 79,7 frente a 64,4 de los clientes retenidos; el 25% de los clientes con menores cargos mensuales que abandonÃ³ pagÃ³ 56,2 frente a 18,3 de los retenidos. En promedio, los cargos mensuales fueron de 64,8, los cargos totales de 2.279,7 y el tenure promedio de 32,4 meses. No se observaron diferencias significativas por gÃ©nero ni por consumo de Streaming TV.

Estos hallazgos, obtenidos a partir de los grÃ¡ficos y mÃ©tricas descriptivas, pueden guiar estrategias de retenciÃ³n, priorizando clientes con mayor riesgo de abandono y ajustando incentivos segÃºn tipo de contrato, tenure, modalidad de pago y cargos mensuales.
